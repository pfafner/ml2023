### *Kernel PCA*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA, KernelPCA

from mpl_toolkits.mplot3d import Axes3D
from matplotlib.ticker import NullFormatter
```

## Generate Half-moon

```python
X, color = datasets.make_moons(n_samples=60, random_state=2021)
```

```python
print(X.shape, color.shape)
```

### PCA

```python
pcamodel = PCA(n_components=2)
pca = pcamodel.fit_transform(X)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(121)
ax.scatter(X[:, 0], X[:, 1], c=color, cmap=plt.cm.Spectral)
ax.set_title("Swiss Roll data")
ax = fig.add_subplot(122)
ax.scatter(pca[:, 0], pca[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
#plt.xticks([]), plt.yticks([])
plt.title('Projected data')
plt.show()
```

### Kernel PCA

```python
kpcamodel = KernelPCA(n_components=2, kernel='rbf', gamma=500., 
                      max_iter=500, random_state=19)
kpca = kpcamodel.fit_transform(X)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(121)
ax.scatter(X[:, 0], X[:, 1], c=color, cmap=plt.cm.Spectral)
ax.set_title("Half-moon data")
ax = fig.add_subplot(122)
ax.scatter(kpca[:, 0], kpca[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
#plt.xticks([]), plt.yticks([])
plt.title('Projected data')
plt.show()
```

### Kernel PCA
### Usamos una función polinomial

```python
kpcamodel = KernelPCA(n_components=2, kernel='poly', gamma=500., degree=3, 
                      max_iter=500, random_state=19)
kpca = kpcamodel.fit_transform(X)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(121)
ax.scatter(X[:, 0], X[:, 1], c=color, cmap=plt.cm.Spectral)
ax.set_title("Half-moon data")
ax = fig.add_subplot(122)
ax.scatter(kpca[:, 0], kpca[:, 1], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
#plt.xticks([]), plt.yticks([])
plt.title('Projected data')
plt.show()
```

```python
kpcamodel = KernelPCA(n_components=2, kernel='poly', gamma=0.5, degree=3, 
                      max_iter=500, random_state=19)
kpca = kpcamodel.fit_transform(X)
```

## Ejemplo con Swiss Roll

```python
X, color = datasets.make_swiss_roll(n_samples=1500, random_state=2021)
```

```python
print(X.shape, color.shape)
```

```python
kpcamodel = KernelPCA(n_components=3, kernel='sigmoid', gamma=100., degree=10, 
                      max_iter=500, random_state=2023)
kpca = kpcamodel.fit_transform(X)
```

```python
fig = plt.figure(figsize=(12,6))
ax = fig.add_subplot(121, projection='3d')
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax.set_title("Swiss Roll data")
ax = fig.add_subplot(122, projection='3d')
ax.scatter(kpca[:, 0], kpca[:, 1], kpca[:, 2], c=color, cmap=plt.cm.Spectral)
plt.axis('tight')
plt.xticks([]), plt.yticks([])
plt.title('Projected data')
plt.show()
```
