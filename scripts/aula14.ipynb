```python
import numpy as np
import matplotlib.pyplot as plt

from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering

from sklearn.datasets import load_iris, make_circles, make_moons, make_blobs

from PIL import Image
```

## Agrupamiento Jer√°rquico

### Burbujas

```python
np.random.seed(2023)
```

```python
data, colors = make_blobs(n_samples=250, n_features=2)
```

```python
print(data.shape, colors.shape)
```

```python
plt.figure()
plt.scatter(data[:,0], data[:,1], c=colors)
plt.show()
```

```python
# par√°metros

k = 3
dist = 'euclidean'
```

```python
# m√©todo = 'single'
model_s = AgglomerativeClustering(n_clusters=k, affinity=dist, linkage='single')
model_s.fit(data)
labels_s = model_s.labels_

# m√©todo = 'complete'
model_c = AgglomerativeClustering(n_clusters=k, affinity=dist, linkage='complete')
model_c.fit(data)
labels_c = model_c.labels_

# m√©todo = 'average'
model_a = AgglomerativeClustering(n_clusters=k, affinity=dist, linkage='average')
model_a.fit(data)
labels_a = model_a.labels_

# m√©todo = 'ward'
model_w = AgglomerativeClustering(n_clusters=k, affinity='euclidean', linkage='ward')
model_w.fit(data)
labels_w = model_w.labels_
```

```python
labels = [labels_s, labels_c, labels_a, labels_w]
models = ['single', 'complete', 'average', 'ward']

plt.figure(figsize=(16,4))
for i in range(0,4):
    plt.subplot(1,4,i+1)
    plt.scatter(data[:,0], data[:,1], c=labels[i])
    plt.title('method = {}'.format(models[i]))
plt.show()
```

```python
labels_a
```

```python
link = linkage(data, metric='euclidean', method='weighted', optimal_ordering=True)
```

```python
plt.figure(figsize=(10,8))
D = dendrogram(link, color_threshold=0)
plt.show()
```

```python
plt.figure(figsize=(10,8))
D = dendrogram(link, color_threshold=3.1)
plt.show()
```

```python
D.keys()
```

```python
print(list(D['color_list']))
```

## Ejemplo con imagen

```python
I = plt.imread('horse.jpg')
I.shape
```

```python
I = Image.open('horse.jpg')
```

```python
factor = 0.25
J = I.resize((int(I.size[0]*factor), int(I.size[1]*factor)), Image.LANCZOS)
```

```python
JJ = np.array(J)
```

```python
JJ.shape
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(I)
plt.subplot(1,2,2)
plt.imshow(JJ)
plt.show()
```

```python
JJ
```

```python
X = JJ.reshape(-1,3)
```

```python
X.shape
```

```python
X[0]
```

```python
k = 2
```

```python
met  = 'average'
dist = 'cosine'
```

```python
model_s = AgglomerativeClustering(n_clusters=k, affinity=dist, linkage=met)
model_s.fit(X)
labels_s = model_s.labels_.reshape(JJ.shape[0], JJ.shape[1])
```

```python
centroids = []
for i in range(0, k):
    centroids.append(JJ[labels_s == i].mean(axis=0).astype(np.uint8))
```

```python
centroids
```

```python
print(model_s.labels_.min(), model_s.labels_.max())
```

```python
Im = np.zeros(JJ.shape).astype(np.uint8)
for i in range(0, k):
    Im[labels_s == i] = centroids[i]

Im = Im.reshape(JJ.shape[0], JJ.shape[1], 3)
```

```python
plt.figure(figsize=(5,4))
plt.imshow(labels_s)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(J)
plt.subplot(1,2,2)
plt.imshow(Im)
plt.show()
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(J)
plt.subplot(1,2,2)
plt.imshow((labels_s==1).astype(np.uint8), cmap='gray')
plt.show()
```

```python
mask = (labels_s==1).astype(np.uint8)
```

```python
M = np.stack([mask, mask, mask], axis=2)
```

```python
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(J)
plt.subplot(1,2,2)
plt.imshow(J*M)
plt.show()
```
